```python
import pandas as pd
from sklearn.metrics import silhouette_score as ss
import matplotlib.pyplot as mp
from sklearn import set_config
import seaborn as sn
from sklearn.pipeline import Pipeline as PP
import numpy as np
from sklearn.cluster import KMeans as KM
set_config(display="diagram")
t=pd.read_excel("cust_data.xlsx")
t.head()
```

```python
t.isnull().sum()
```

```python
t.dropna(inplace=True)
```

```python
#t.Gender=np.where(t.Gender=="M",1,0)
```

```python
t.head()
```

```python
t.duplicated().sum()
```

```python
t=t[t.Orders!=0]
```

```python
x=t.drop(["Cust_ID","Gender","Orders"],axis=1)
x
```

```python
lo=[KM(n_clusters=i).fit(x).inertia_ for i in range(1,10)]
lo
```

```python
mp.plot(range(1,10),lo,"rd-")
mp.bar(range(1,10),lo,color=sn.color_palette())
mp.grid();
mp.xlabel("Elbow-method");
```

```python
lo=[ss(x,KM(n_clusters=i).fit(x).labels_) for i in range(2,10)]
mp.plot(range(2,10),lo,"ro--")
mp.grid()
mp.xlabel("Silhouette score");
```

```python
pp=PP(
    [
        ("KM",KM(n_clusters=2))
]
)
```

```python
y1=pp.fit_predict(x)
y1
```

```python
t["y"]=y1
```

```python
t.groupby("y")["Orders"].sum()
```

```python
pd.crosstab(t.y,[t.Gender]).plot(kind="bar")
mp.legend(loc="upper left");
```

```python
pd.crosstab(t.y,[t.Orders]).plot(kind="bar")
mp.legend(loc="upper right");
```
